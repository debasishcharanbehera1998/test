```python
a = ['Apple','Orange','Grapes','Pineapple']
```

```python
b = [200, 400, 500, 600]
```

```python
[i for i in a if "APPLE" in i.upper()]
```

```python
[i for i in a if "APPLE" not in i.upper()]
```

```python
df1 = pd.DataFrame(zip(a, b), columns=["Fruits", "Quantity"], index = [1, 2, 3, 4])
df1
```

```python
df1["Price"] = df1.apply(lambda x : 10 if x["Fruits"].upper() == "APPLE" else 20 if x["Fruits"].upper() == "ORANGE" else 30 if x["Fruits"].upper() == "GRAPES" else 40 if x["Fruits"].upper() == "PINEAPPLE" else 0, axis = 1)
```

```python
df1
```

```python
df1.rename(columns={"Price" : "Cost_per_piece"},inplace = True)
```

```python
df1
```

```python
df1["Overall_Price"] = df1.apply(lambda x : x["Quantity"] * x["Cost_per_piece"], axis = 1)
```

```python
df1
```

```python
# Sorting on basis of Overall Price

df1.sort_values(by = "Overall_Price", ascending=False, inplace=True)
```

```python
df1
```

```python
df1[df1["Overall_Price"] == max(df1["Overall_Price"])].reset_index()
```

```python
elements = a.copy()
SP = [30000, 20000, 10000, 2000]
Selling = pd.DataFrame(zip(elements, SP), columns=["Fruits", "Selling_Price"])
```

```python
Selling
```

```python
df1
```

```python
df1 = df1.merge(right = Selling, how = "left", left_on="Fruits", right_on="Fruits")
```

```python
df1["Profit_loss_Amt"] = df1["Selling_Price"]-df1["Overall_Price"]
```

```python
df1
```

```python
df1.rename(columns={"Overall_Price": "Cost_Price"}, inplace=True)
```

```python
df1
```

```python
df1.drop(columns="Quantity", inplace=True)
```

```python
df1
```

```python
df1["Profit&loss_pct"] = df1.apply(lambda x : str(x["Profit_loss_Amt"] * 100 /x["Cost_Price"]) + " %", axis = 1 )
```

```python
df1
```

```python
df1["Loss_or_profit"] = df1.apply(lambda x: "Loss" if x["Profit_loss_Amt"] < 0 else "Profit", axis = 1)
```

```python
df1
```

```python
df1["Loss_or_profit"] = df1.apply(lambda x : x["Loss_or_profit"].upper(), axis = 1)
```

```python
# FInbonacci Serties


def fibonaci(n):
    F  = [0, 1]
    for i in range(2, n+1):
        F.append(F[i-1] + F[i-2])

    return F

fibonaci(5)
```
